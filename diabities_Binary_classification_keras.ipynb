### Required LIbraries

```python
# libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# import modules
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math  import sqrt

# import keras specific
import keras
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.utils import to_categorical 
```

```python
def plot_loss(history):
    
    "pass history as Data Frame "
    plt.figure(figsize=(20,10))
    try:
        plt.plot(history["loss"], label = "loss")
        plt.plot(history["val_loss"], label = "val_loss")
    except:
        pass
    try:
        plt.plot(history["accuracy"], label = "Accuracy")
        plt.plot(history["val_accuracy"], label = "val_accuracy")
    except:
        pass
    
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy ")
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Load Data sets

```python
df = pd.read_csv("datas/diabetes.csv")
print(df.shape)
df.describe().T
```

```python
df.head()
```

```python
# our target classification column is Outcome which is True or False 1 & 0
target_column = ["Outcome"]
predictors = list(set(list(df.columns)) - set(target_column))
predictors
```

```python
df[predictors] = df[predictors] / df[predictors].max() # Doing Normalization this can done directly using sklearn
df.describe().T
```

## Creating the Training and Test DataSets

```python
df.head()
```

```python
x = df[predictors].values # changing to numpy 
y = df[target_column].values

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
print(x_train.shape, x_test.shape)
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
y_train # For Deep learning need to change to categorical field
# also can do with pandas get_dummies
```

```python
count_classes = y_test.shape[1]
print(count_classes)
```

```python
y_test[:5]
```

### Define, Compile, and Fit the Keras Classification Model

```python
model = Sequential()
model.add(Dense(500, activation="relu", input_dim = 8))
model.add(Dense(100, activation ="relu"))
model.add(Dense(50, activation = "relu"))
model.add(Dense(2, activation = "softmax")) # Soft max Show the probability of High Risk & Low Risk
```

```python
model.summary()
```

```python
# Compile the Model
model.compile(optimizer="adam",
             loss = "categorical_crossentropy",
             metrics = ["accuracy"])
```

```python
history = model.fit(x_train, y_train, epochs=50, validation_split=0.2)
```

```python
pd.DataFrame(history.history).head()
```

```python
plot_loss(history.history)
```

## Predict on the Test Data and Compute Evaluation Metrics

```python
pred_train = model.predict(x_train)
scores = model.evaluate(x_train,y_train, verbose = 0)
print("Accuracy on training data ", scores[1], "Error in Training data is ", 1 - scores[1]) # 100 is full

pred_test = model.predict(x_test)
scores2 = model.evaluate(x_test, y_test, verbose=1)
print("Accuracy on test data ", scores[1], "Error in Training data is ", 1 - scores[1])
```
