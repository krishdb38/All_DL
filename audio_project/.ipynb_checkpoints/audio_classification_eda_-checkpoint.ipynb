## Extract Features
**Mel-Frequency Cepstral Coefficients(MFCC)**
Here we will be using Mel-Frequency Cepstral Coefficients (MFCC) from the audio samples. <br>
The MFCC summarises the frequency distribution across the window size, so  it is possible to analyse both the frequency and time characteristics of the sound.
These representations will allow us to identify features for classification

```python
mfcss = librosa.feature.mfcc(y = librosa_audio_data, sr = librosa_sample_rate, n_mfcc = 40)
```

```python
##

def features_extractor(file):
    audio,sample_rate = librosa.load(file_name, res_type = "kaiser_fast")
    mfccs_features = librosa.feature.mfcc(y = audio, sr = sample_rate, n_mfcc = 40)
    #
    mfccs_scaled_eatures = np.mean(mfccs_features.T, axis = 0)
    
    return mfccs_scaled_eatures
```

```python
audio_dataset_path = "../data_sets/Urban_sound_8k/audio/"
```

```python
features_extractor(filename)
```

```python
# Now we iterate through every audio file and extract features

# using Mel-Frequency Cepstral Coefficients

extracted_features = []

for index_num, row in tqdm(metadata.iterrows()):
    #file_name = os.path.join(os.path.abspath(audio_dataset_path), 
    #                         "fold" + str(row["fold"]) + "\\" , str(row["slice_file_name"]))
    # Since Jupyter might not have full path authority
    file_name = os.path.join(audio_dataset_path ,"fold"+ str(row["fold"]),str((row["slice_file_name"])) )
    
    final_class_labels = row["class"]
    #print(file_name)
    data = features_extractor(file_name)
    extracted_features.append([data, final_class_labels])
```

```python
# COnverting extracted_features to Pandas dataFrame
extracted_features_df = pd.DataFrame(extracted_features, columns = ["feature", "class"])
extracted_features_df.head()
```

```python
#extracted_features_df.to_csv("../data_sets/audio_classification.csv", index = None)
```

```python
extracted_features_df.shape
```

```python
# Split the dataset into independent and dependent dataset
x = np.array(extracted_features_df["feature"].tolist())
y = np.array(extracted_features_df["class"].tolist())


## THis can be done simply with 
# x = extracted_features_df["feature"].values()
```

```python
x.shape
```

```python
y
```

```python
## Label Encoding
##  y = np.array(pd.get_dummies(y))
## 

from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
y = to_categorical(labelencoder.fit_transform(y))
```

```python
y
```

```python
##  Train Test Split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### Model Creation

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.optimizers import Adam
from sklearn import metrics
```

```python
# Unique value in y, Category to predict
y.shape
```

```python
# There are 10 Categories
print(x_train.shape)
x_train[1]
```

```python
# Output level
num_labels = y.shape[1] # 10 
```

```python
model = Sequential()
model.add(Dense(100, input_shape = (40,))) #input size 40
model.add(Activation ("relu"))
model.add(Dropout(0.5))

# Second Layer
model.add(Dense(200))
model.add(Activation("relu"))
model.add(Dropout(0.5))

# Third Layer
model.add(Dense(100))
model.add(Activation("relu"))
model.add(Dropout(0.5))

### final layer
model.add(Dense(num_labels))
model.add(Activation("softmax"))
```

```python
model.summary()
```

```python
model.compile(loss = "categorical_crossentropy", metrics = ["accuracy"], optimizer = "adam")
```

```python
# Compile the mode
from tensorflow.keras.callbacks import ModelCheckpoint
from datetime import datetime

num_epochs = 100
num_batch_size = 32

checkpointer = ModelCheckpoint(filepath = "../data_sets/Urban_sound_8k/audio_classification_urban.hdf", verbose = 1, 
                              save_best_only = True)
```

```python
start = datetime.now()
model.fit(x_train, y_train, batch_size = num_batch_size, epochs = num_epochs, validation_data = (x_test, y_test),
          callbacks = [checkpointer], verbose = 1, validation_freq=10)
duration = datetime.now() - start
print("Training completed in time ", duration)
```

```python
test_accuracy = model.evaluate(x_test, y_test, verbose = 0)
print(test_accuracy)
```
