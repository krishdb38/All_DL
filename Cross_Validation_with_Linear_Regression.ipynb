```python
# ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
# these columns have only binary field so change to 0 & 1 
binary_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

def binary_map(x):
    return x.map({"yes":1, "no":0})

# Applying the function to the variables list
housing[binary_cols] = housing[binary_cols].apply(binary_map)
housing.head()
```

```python
## in the above table last column furnishingstatus
# also 
housing = pd.get_dummies(housing, columns=["furnishingstatus"], drop_first=True,)
housing.head()
```
